```python
import pandas as pd
import numpy as np
from lib import *

file = "../00_data/sample_10_22_matt.csv"

general_movies = pd.read_csv(file)
```

```python
general_movies.head(10)
```

```python
critics = {}
movies_not_found = []
counter = 0

for row in general_movies.iterrows():
    try:
        soup = getting_soup(
            row[1]["Movie Title"], time_waiting=np.random.randint(5, 15)
        )
        similiar_movies = getting_movies_from_soup(soup)
        # Filtering by year
        similiar_movies = {
            k: v
            for k, v in similiar_movies.items()
            if v["release_year"] == str(row[1]["Year"])
        }
        # movie in rotten tomatoes format
        movie_rt_format = list(similiar_movies.keys())[0]
        critics[row[1]["Movie Title"]] = {
            "formatted_name": movie_rt_format,
            "top_critics": getting_critics(movie_rt_format),
        }

    except:
        movies_not_found.append(row[1]["Movie Title"])
        print(f"Error with {row[1]['Movie Title']}")
        continue

    counter += 1
    if counter % 10 == 0:
        print(f"{counter} movies processed")
        time_sleeping = np.random.randint(5, 7)
        print(f"sleeping :{time_sleeping} seconds")
        time.sleep(np.random.randint(5, 15))
```

```python
# saving the critics
critics_df = pd.DataFrame(
    {
        "movie": list(critics.keys()),
        "formatted_name": list(map(lambda x: x["formatted_name"], critics.values())),
        "top_critics": list(map(lambda x: x["top_critics"], critics.values())),
    }
)

# merging the critics with the movies
general_movies = general_movies.merge(
    critics_df, left_on="Movie Title", right_on="movie", how="outer", indicator=True
)
```

```python
general_movies.to_csv(file, index=False)
```
